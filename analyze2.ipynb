```python
import pandas as pd
from scipy.stats import spearmanr
import matplotlib.pyplot as plt

# 데이터 로드
salary_rank_data = pd.read_csv('Salary_by_Team_Rank.csv')

# 스피어만 상관계수 계산
correlation, p_value = spearmanr(salary_rank_data["Salary (£)"], salary_rank_data["Final Rank"])

# 상관계수와 p-value 출력
print(f"Spearman Correlation: {correlation:.2f}")
print(f"P-value: {p_value:.2e}")

# 연봉과 순위의 상관관계 시각화
plt.figure(figsize=(12, 6))
plt.scatter(salary_rank_data["Salary (£)"], salary_rank_data["Final Rank"], alpha=0.8, edgecolor="k")
plt.title(f"Salary vs Final Rank (Spearman Correlation: {correlation:.2f})", fontsize=14)
plt.xlabel("Salary (£)", fontsize=12)
plt.ylabel("Final Rank (Lower is Better)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.gca().invert_yaxis()  # 순위는 낮을수록 좋은 순위를 반영하므로 y축 반전
plt.show()
```
