```python
import numpy as np

# Wage 열에서 숫자로 변환하는 함수 정의
def parse_wage(value):
    if isinstance(value, str) and 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1_000
    elif isinstance(value, str) and 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1_000_000
    else:
        return np.nan

# Wage 열 숫자 변환
df['Wage (Numeric)'] = df['Wage'].apply(parse_wage)

# Overall rating 숫자 변환
df['Overall rating'] = pd.to_numeric(df['Overall rating'], errors='coerce')

# 두 변수 간 상관계수 계산
correlation = df[['Overall rating', 'Wage (Numeric)']].corr()

# 결과 출력
print("Overall rating과 연봉 간의 상관계수:")
print(correlation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 산점도를 통한 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Overall rating', y='Wage (Numeric)', color='blue', alpha=0.7)

# 그래프 제목 및 축 레이블 추가
plt.title('Overall Rating vs. Wage', fontsize=16)
plt.xlabel('Overall Rating', fontsize=14)
plt.ylabel('Wage (in Euros)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

# 그래프 표시
plt.show()
```

```python
# 골키퍼 관련 열 선택
gk_attributes = ["GK Diving", "GK Handling", "GK Kicking", "GK Positioning", "GK Reflexes"]

# 숫자로 변환하여 유효한 데이터만 필터링
gk_data_cleaned = df[gk_attributes].apply(pd.to_numeric, errors='coerce')

# 골키퍼 능력치의 평균 계산
gk_mean_cleaned = gk_data_cleaned.mean()

# 능력치 평균을 바 차트로 시각화
plt.figure(figsize=(10, 6))
gk_mean_cleaned.plot(kind='bar', color='skyblue', alpha=0.8)
plt.title('Cleaned Average GK Attributes', fontsize=16)
plt.ylabel('Average Score', fontsize=14)
plt.xlabel('GK Attributes', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()

# 각 골키퍼의 평균 능력치 계산
gk_data_cleaned['GK Average'] = gk_data_cleaned.mean(axis=1)

# 상위 5명 골키퍼 추출
top_gk = df[['name']].join(gk_data_cleaned[['GK Average']]).sort_values(by='GK Average', ascending=False).head()

# 상위 5명의 골키퍼 출력
print(top_gk)
```

```python
# 연봉 데이터를 숫자형으로 변환하는 함수
def parse_wage(value):
    if isinstance(value, str) and 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1_000
    elif isinstance(value, str) and 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1_000_000
    else:
        return None

# "Best position"이 "GK"인 선수들 필터링
gk_only_df = df[df['Best position'] == 'GK']

# 골키퍼 관련 열 필터링
gk_attributes = ["GK Diving", "GK Handling", "GK Kicking", "GK Positioning", "GK Reflexes"]
gk_only_data = gk_only_df[gk_attributes].copy()

# 연봉 데이터를 숫자로 변환
gk_only_df.loc[:, 'Wage (Numeric)'] = gk_only_df['Wage'].apply(parse_wage)

# 골키퍼 평균 능력치 계산
gk_only_data['GK Average'] = gk_only_data.select_dtypes(include='number').mean(axis=1)

# 유효 데이터 필터링
valid_data = gk_only_data[['GK Average']].join(gk_only_df[['Wage (Numeric)']]).dropna()

# 골키퍼 평균 능력치와 연봉 간의 상관관계 계산
correlation_gk_only = valid_data.corr()

# 시각화
plt.figure(figsize=(10, 6))
plt.scatter(valid_data['GK Average'], valid_data['Wage (Numeric)'], alpha=0.7, color='green')
plt.title('GK Average vs Wage (GK only)', fontsize=16)
plt.xlabel('GK Average (Mean of GK Attributes)', fontsize=14)
plt.ylabel('Wage (in Euros)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 상관관계 결과 출력
print(correlation_gk_only)
```

```python
# 연봉 데이터를 숫자형으로 변환하는 함수 정의
def parse_wage(value):
    if isinstance(value, str) and 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1_000
    elif isinstance(value, str) and 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1_000_000
    else:
        return None

# Wage 데이터를 숫자로 변환하여 새로운 열 생성
df['Wage (Numeric)'] = df['Wage'].apply(parse_wage)

# 중앙 미드필더(CM) 선수들 필터링
cm_df = df[df['Best position'] == 'CM'].copy()  # 복사본 생성

# 패스 데이터를 숫자로 변환하는 함수 정의
def clean_numeric(value):
    try:
        return float(value.split('+')[0])  # '+' 기호 제거 후 숫자로 변환
    except:
        return None

# 패스 관련 열에 변환 적용 (loc 사용)
passing_attributes = ['Short passing', 'Long passing', 'Vision']
for attr in passing_attributes:
    cm_df.loc[:, attr] = cm_df[attr].apply(clean_numeric)

# 변환 후 상관관계 계산
cm_passing_wage_corr_cleaned = cm_df[passing_attributes + ['Wage (Numeric)']].dropna().corr()
passing_wage_correlations_cleaned = cm_passing_wage_corr_cleaned['Wage (Numeric)'].sort_values(ascending=False)

# 상관관계 결과 출력
print("패스 능력과 연봉 간의 상관관계:")
print(passing_wage_correlations_cleaned)
```

```python
# 연봉 데이터를 숫자형으로 변환
def parse_wage(value):
    try:
        value = value.split('+')[0]  # '+' 기호 이전 부분만 사용
        if 'K' in value:
            return float(value.replace('€', '').replace('K', '')) * 1_000
        elif 'M' in value:
            return float(value.replace('€', '').replace('M', '')) * 1_000_000
        else:
            return float(value.replace('€', '').replace(',', ''))
    except:
        return None

# 공격 능력 데이터를 숫자로 변환
def clean_numeric(value):
    try:
        return float(value.split('+')[0])  # '+' 기호 이전 숫자만 추출
    except:
        return None

# 공격수(ST) 필터링
st_df = df[df['Best position'].str.contains('ST', na=False)]

# 공격 능력의 주요 열과 연봉 데이터를 포함하여 결측값 평균으로 대체
attacker_attributes_corrected = [
    'Finishing', 'Att. Position', 'Volleys', 'Heading accuracy',
    'Shot power', 'Long shots'
]

# 데이터 복사 및 변환
st_filled_data = st_df[attacker_attributes_corrected + ['Wage']].copy()
st_filled_data['Wage (Numeric)'] = st_filled_data['Wage'].apply(parse_wage)

# 주요 능력치를 숫자로 변환
for attr in attacker_attributes_corrected:
    st_filled_data[attr] = st_filled_data[attr].apply(clean_numeric)

# 결측값 처리: 주요 능력치는 0으로 대체
st_filled_data[attacker_attributes_corrected] = st_filled_data[attacker_attributes_corrected].fillna(0)

# 공격 능력 총합 계산
st_filled_data['Total Attacking Ability'] = st_filled_data[attacker_attributes_corrected].sum(axis=1)

# 총합과 연봉 간 상관관계 계산
valid_st_data = st_filled_data[['Total Attacking Ability', 'Wage (Numeric)']].dropna()
total_attacking_corr = valid_st_data.corr()
attacking_wage_corr = total_attacking_corr.loc['Total Attacking Ability', 'Wage (Numeric)']

# 상관관계 출력
print("Total Attacking Ability와 Wage 간의 상관관계:", attacking_wage_corr)
```

```python
import pandas as pd
from scipy.stats import spearmanr
import matplotlib.pyplot as plt

# 데이터 로드
salary_rank_data = pd.read_csv('Salary_by_Team_Rank.csv')

# 스피어만 상관계수 계산
correlation, p_value = spearmanr(salary_rank_data["Salary (£)"], salary_rank_data["Final Rank"])

# 상관계수와 p-value 출력
print(f"Spearman Correlation: {correlation:.2f}")
print(f"P-value: {p_value:.2e}")

# 연봉과 순위의 상관관계 시각화
plt.figure(figsize=(12, 6))
plt.scatter(salary_rank_data["Salary (£)"], salary_rank_data["Final Rank"], alpha=0.8, edgecolor="k")
plt.title(f"Salary vs Final Rank (Spearman Correlation: {correlation:.2f})", fontsize=14)
plt.xlabel("Salary (£)", fontsize=12)
plt.ylabel("Final Rank (Lower is Better)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.gca().invert_yaxis()  # 순위는 낮을수록 좋은 순위를 반영하므로 y축 반전
plt.show()
```
